```python
import numpy as np
import matplotlib.pyplot as plt

# 한글 처리
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 평균, 표준편차 설정
mu, sigma = 0, 1

# 표준 정규분포에서 1,000,000개의 샘플 생성
samples = np.random.normal(mu, sigma, 1000000)

# 그래프 그리기
plt.hist(samples, bins=50, density=True, alpha=0.2, color='skyblue', label='표준 정규분포')
plt.title('표준 정규분포')
plt.xlabel("x")
plt.ylabel("y")

# 표준 정규 분포의 확률 밀도 함수 그리기
x = np.linspace(-5, 5, 100)
pdf = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2))
plt.plot(x, pdf, 'b-.', linewidth=2, label='확률 밀도 함수')

plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 정규분포를 따르는 1000개의 난수 생성 (평균=0, 표준편차=1)
data = np.random.randn(100000)

# 2. 히스토그램 그리기
plt.hist(data, bins=100, edgecolor='black')  # bins=30: 구간 수 조절
plt.title("Histogram of 1000 Normally Distributed Values")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
